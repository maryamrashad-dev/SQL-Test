# Container Data Types

## List 
### ordered, with indices starting from 0

```python
name_parts = ['My','Name','is','This']
```

```python
type(name_parts)
```

```python
len(name_parts)
```

```python
# first element
name_parts[0]
```

```python
# last element
print(name_parts[3])
print(name_parts[len(name_parts)-1])
print(name_parts[-1])
```

```python
# first two element
print(name_parts[0:2])
print(name_parts[:2])
```

```python
# second to last
name_parts[-2]
```

```python
# add element to end of list
name_parts.append('not')
name_parts
```

```python
name_parts.append('that')
name_parts
```

```python
# add element at specific index
name_parts.insert(3,'Maryam')
name_parts
```

```python
# remove a value
name_parts.pop()
name_parts
```

```python
name_parts.pop(4)
name_parts
```

```python
name_parts.remove('not')
name_parts
```
